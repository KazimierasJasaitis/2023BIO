```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate

def sin_formula(alpha,x):
  return np.sin(alpha * x + np.pi)

def trapezoid(i,alpha):
  S = 0
  for i in range(n):
    x_left = h * i
    x_right = h * ( i + 1 )
    S += (sin_formula(alpha,x_left) + sin_formula(alpha,x_right)) / 2 * h
  return S

def simpson(i, alpha):
  S = 0
  for i in range(n//2):
    x_left = h * (2*i)
    x_middle = h * (2*i + 1)
    x_right = h * (2*i + 2)
    S += (sin_formula(alpha, x_left) + 4*sin_formula(alpha, x_middle) + sin_formula(alpha, x_right)) * h / 3
  return S


a = 0
b = 1
n = 2
h = (b-a)/n
Ss = [] #Simpson sum
St = [] #trapezoid sum
Sc = [] #control integration by scipy.integrate sum

alpha_values = np.arange(0.01, 1.01, 0.01)

for alpha in alpha_values:
  Ss.append(simpson(i,alpha))
  St.append(trapezoid(i,alpha))
  Sc.append(integrate.quad(sin_formula, a, b, args=(alpha,))[0])

plt.figure(figsize=(12, 6))

diagram = plt.plot(Ss, label="Ss", color='yellow', alpha=0.5)
diagram = plt.plot(St, label="St", color='cyan', alpha=0.3)
diagram = plt.plot(Sc, label="Sc", color='magenta', alpha=0.3)

plt.legend(loc="upper right")
plt.gca().set_facecolor('black')

print(Ss)
print(St)
print(Sc)
```
